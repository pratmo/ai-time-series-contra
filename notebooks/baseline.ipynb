```python
# ==============================================================
# Rolling-origin CV for CNN-LSTM with honest recursive evaluation
# ==============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -----------------------
# 0) Reproducibility
# -----------------------
np.random.seed(42)
tf.random.set_seed(42)

# -----------------------
# 1) Load & sort data
# -----------------------
CSV_PATH = "/kaggle/input/cnn-lstm-v2/df_bax_cleaned_till_outliers.csv"   # <-- change if needed
df = pd.read_csv(CSV_PATH)
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").set_index("Date")

# Use only the target column "Price"
y = df["Price"].astype("float32").values.reshape(-1, 1)
dates = df.index

# -----------------------
# 2) Global split: 80% train_block, 20% test_block (untouched for model selection)
# -----------------------
n = len(y)
split_80 = int(0.80 * n)
y_train_block = y[:split_80]       # will be used for CV + final training
y_test_block  = y[split_80:]       # final test only (never used in CV)
dates_test    = dates[split_80:]

# -----------------------
# 3) Inside the 80% block, break into 4 equal segments (A,B,C,D)
#    With 4 segments, rolling-origin gives 3 folds:
#    Fold1: Train=A,      Val=B
#    Fold2: Train=A+B,    Val=C
#    Fold3: Train=A+B+C,  Val=D
# -----------------------
def split_into_four_segments(arr):
    seg_len = len(arr) // 4
    A = arr[:seg_len]
    B = arr[seg_len:2*seg_len]
    C = arr[2*seg_len:3*seg_len]
    D = arr[3*seg_len:]  # may be a bit longer if not divisible by 4
    return A, B, C, D

A, B, C, D = split_into_four_segments(y_train_block)

folds = [
    # (train_slice, val_slice)
    (np.concatenate([A]), np.concatenate([B])),
    (np.concatenate([A, B]), np.concatenate([C])),
    (np.concatenate([A, B, C]), np.concatenate([D])),
]

# -----------------------
# 4) Helper: build sequences X,y from a 1D series (already scaled)
#    We build ONLY within the passed array to avoid leakage.
# -----------------------
def make_sequences(arr_scaled, time_steps):
    X, y = [], []
    for i in range(len(arr_scaled) - time_steps):
        X.append(arr_scaled[i:i+time_steps, 0])     # past window
        y.append(arr_scaled[i+time_steps, 0])       # next step
    X = np.array(X)[..., np.newaxis]                # shape: (samples, time, 1)
    y = np.array(y)                                 # shape: (samples,)
    return X, y

# -----------------------
# 5) Helper: build a CNN-LSTM model from hyperparameters
# -----------------------
def build_model(hp, time_steps):
    model = Sequential()
    model.add(Conv1D(
        filters=hp["conv_filters"],
        kernel_size=hp["kernel_size"],
        activation=hp["conv_activation"],
        input_shape=(time_steps, 1)
    ))
    model.add(MaxPooling1D(pool_size=hp["pool_size"]))
    model.add(LSTM(hp["lstm_units"], activation=hp["lstm_activation"]))
    model.add(Dropout(hp["dropout"]))
    model.add(Dense(hp["dense_units"]))

    # Optimizer (allow custom LR if provided)
    if hp.get("learning_rate") is not None:
        opt = Adam(learning_rate=hp["learning_rate"])
    else:
        opt = Adam()
    model.compile(optimizer=opt, loss="mse")
    return model

# -----------------------
# 6) Metrics helper
# -----------------------
def compute_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return {"MAE": mae, "RMSE": rmse, "MAPE": mape}

# -----------------------
# 7) Recursive forecast on a target block (val or test)
#    - scaler is fitted on the TRAIN slice only
#    - start_window: last TIME_STEPS points from TRAIN (scaled)
#    - then forecast len(target_block) steps, feeding preds back
# -----------------------


def recursive_forecast(model, scaler, train_block_raw, target_block_raw, time_steps, horizon=15):
    """
    Make an honest recursive multi-step forecast limited to 'horizon' steps.
    - scaler: will be fit on TRAIN only (no leakage).
    - start window: last 'time_steps' points from TRAIN (scaled).
    - we then forecast up to min(horizon, len(target_block_raw)) steps.
    """
    # Fit scaler on TRAIN only
    train_scaled = scaler.fit_transform(train_block_raw.copy())
    target_scaled = scaler.transform(target_block_raw.copy())

    # Initial window from the end of TRAIN
    if len(train_scaled) < time_steps:
        raise ValueError("Train slice shorter than TIME_STEPS.")
    window = train_scaled[-time_steps:, 0].copy()  # shape: (time_steps,)

    # Forecast only up to 'horizon' steps (or less if target shorter)
    steps = min(horizon, len(target_scaled))
    preds_scaled = []
    for _ in range(steps):
        x_in = window.reshape(1, time_steps, 1)
        next_scaled = model.predict(x_in, verbose=0)[0, 0]
        preds_scaled.append(next_scaled)
        # slide window: drop oldest, append prediction
        window = np.roll(window, -1)
        window[-1] = next_scaled

    # Inverse-transform predictions and the corresponding actuals slice
    preds = scaler.inverse_transform(np.array(preds_scaled).reshape(-1, 1)).ravel()
    actual = target_block_raw[:steps].ravel()
    return preds, actual


# -----------------------
# 8) Train-on-slice, validate on next slice (1 fold)
#    Returns validation metrics and the trained model (best weights via ES)
# -----------------------
def train_and_validate_one_fold(train_raw, val_raw, hp):
    scaler = MinMaxScaler((0,1))
    train_scaled = scaler.fit_transform(train_raw.copy())

    X_train, y_train = make_sequences(train_scaled, hp["TIME_STEPS"])

    es = EarlyStopping(monitor="loss", patience=hp["PATIENCE"], restore_best_weights=True, verbose=0)
    model = build_model(hp, hp["TIME_STEPS"])
    model.fit(
        X_train, y_train,
        epochs=hp["EPOCHS"],
        batch_size=hp["BATCH_SIZE"],
        verbose=0,
        callbacks=[es]
    )

    # ⬇️ Forecast only the next 15 days of the validation slice
    preds, actual = recursive_forecast(
        model, MinMaxScaler((0,1)),
        train_block_raw=train_raw,
        target_block_raw=val_raw,
        time_steps=hp["TIME_STEPS"],
        horizon=15                 # <— limit to 15
    )
    val_metrics = compute_metrics(actual, preds)
    return val_metrics, model


# -----------------------
# 9) Define your initial hyperparameters AND an optional small grid to try
# -----------------------
base_hp = {
    "TIME_STEPS": 60,
    "BATCH_SIZE": 32,
    "EPOCHS": 200,
    "PATIENCE": 10,

    "conv_filters": 64,
    "kernel_size": 2,
    "pool_size": 2,
    "conv_activation": "relu",

    "lstm_units": 64,
    "lstm_activation": "tanh",
    "dropout": 0.2,

    "dense_units": 1,
    "learning_rate": None,   # use Adam default
}

# A small, safe grid to start (change as you like)
grid = [
    {**base_hp},
    {**base_hp, "TIME_STEPS": 45},
    {**base_hp, "TIME_STEPS": 75},
    {**base_hp, "lstm_units": 96},
    {**base_hp, "conv_filters": 96},
    {**base_hp, "dropout": 0.3},
    {**base_hp, "learning_rate": 0.0005},
]

# -----------------------
# 10) Run rolling-origin CV for each hyperparameter setting
#      Folds: (A->B), (A+B->C), (A+B+C->D)
# -----------------------
def run_cv_and_select_best(grid):
    results = []
    for i, hp in enumerate(grid):
        fold_metrics = []
        for (train_raw, val_raw) in folds:
            m, _ = train_and_validate_one_fold(train_raw, val_raw, hp)
            fold_metrics.append(m["RMSE"])  # use RMSE to rank

        avg_rmse = float(np.mean(fold_metrics))
        results.append({"hp_index": i, "avg_val_rmse": avg_rmse, "hp": hp})

        print(f"[Setting {i}] RMSE per fold: {np.round(fold_metrics, 4)} | Avg RMSE: {avg_rmse:.4f}")

    # Pick hyperparameters with the lowest average validation RMSE
    best = min(results, key=lambda d: d["avg_val_rmse"])
    print("\nBest by avg validation RMSE:", best["hp"])
    return best

best_choice = run_cv_and_select_best(grid)
best_hp = best_choice["hp"]

# -----------------------
# 11) Train FINAL model on full 0–80% with best HPs, test on 80–100%
# -----------------------
def train_final_and_test(train_block_raw, test_block_raw, hp):
    scaler = MinMaxScaler((0,1))
    train_scaled = scaler.fit_transform(train_block_raw.copy())
    X_train, y_train = make_sequences(train_scaled, hp["TIME_STEPS"])

    es = EarlyStopping(monitor="loss", patience=hp["PATIENCE"], restore_best_weights=True, verbose=0)
    model = build_model(hp, hp["TIME_STEPS"])
    model.fit(
        X_train, y_train,
        epochs=hp["EPOCHS"],
        batch_size=hp["BATCH_SIZE"],
        verbose=0,
        callbacks=[es]
    )

    # ⬇️ Only forecast the next 15 days on the untouched test block
    preds, actual = recursive_forecast(
        model, MinMaxScaler((0,1)),
        train_block_raw=train_block_raw,
        target_block_raw=test_block_raw,
        time_steps=hp["TIME_STEPS"],
        horizon=15                 # <— limit to 15
    )
    test_metrics = compute_metrics(actual, preds)
    return model, preds, actual, test_metrics


final_model, test_preds, test_actual, test_metrics = train_final_and_test(y_train_block, y_test_block, best_hp)

print("\n=== Final Test Metrics (80–100, recursive) ===")
for k, v in test_metrics.items():
    print(f"{k}: {v:.4f}")

# -----------------------
# 12) Naive baseline on test (y_hat_t = y_{t-1})
# -----------------------
naive_preds = []
prev = y_train_block[-1, 0]  # last training actual as yesterday for first test point
for t in range(len(y_test_block)):
    naive_preds.append(prev)
    prev = y_test_block[t, 0]  # update with actual (definition of naive-1 baseline)
naive_preds = np.array(naive_preds)

naive_metrics = compute_metrics(y_test_block.ravel(), naive_preds)
print("\n=== Naive-1 Baseline on Test ===")
for k, v in naive_metrics.items():
    print(f"{k}: {v:.4f}")

# -----------------------
# 13) Plot: Test actual vs final recursive forecast (and baseline)
# -----------------------
# dates_test corresponds to the full 20% test span; show only first 15 forecasted points
h = 15
plt.figure(figsize=(14,6))
plt.plot(dates_test[:h], y_test_block[:h].ravel(), label="Actual (Test, 15d)")
plt.plot(dates_test[:h], test_preds[:h], label="Forecast (15d)")
plt.plot(dates_test[:h], naive_preds[:h], label="Naive-1 baseline", linestyle="--")
plt.title("BAX – 15-Day Test Forecast")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()

```
